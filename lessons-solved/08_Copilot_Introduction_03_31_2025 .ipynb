# Basic Usage

## 1. Code Completion

Copilot provides suggestions as you type. Let's try some examples:

1. Start typing a function definition:
```python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
```

2. Copilot will suggest the implementation. Press Tab to accept or keep typing for different suggestions.

````python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
    if height <= 0:
        raise ValueError("Height must be greater than zero.")
    if weight <= 0:
        raise ValueError("Weight must be greater than zero.")
    bmi = weight / (height ** 2)
    return round(bmi, 2)

def bmi_category(bmi):
    # Determine the BMI category based on the calculated BMI
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obesity"

def main():
    try: 
        weight = float(input("Enter your weight in kilograms: "))
        height = float(input("Enter your height in meters: "))
        bmi = calculate_bmi(weight, height)
        category = bmi_category(bmi)
        print(f"Your BMI is: {bmi}")
        print(f"You are classified as: {category}")
    except ValueError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
````

````python
def calculate_dosage(weight, age, condition):
    """Calculate medication dosage based on patient parameters."""
    # Check if weight or age is invalid (non-positive values)
    if weight <= 0 or age < 0:
        raise ValueError("Weight and age must be positive")
    
    # Calculate the base dosage as weight multiplied by 2
    base_dosage = weight * 2
    
    # If the patient is a child (age < 12), reduce the dosage by 25%
    if age < 12:
        base_dosage *= 0.75
    
    # If the condition is severe, increase the dosage by 50%
    if condition == "severe":
        base_dosage *= 1.5
    
    # Return the final dosage rounded to 2 decimal places
    return round(base_dosage, 2)

# Write pytest test cases for the above function
# Include tests for normal cases, edge cases, and error conditions
def test_calculate_dosage():
    # Test normal case: weight=50, age=10, condition="normal"
    assert calculate_dosage(50, 10, "normal") == 100.0
    
    # Test normal case: weight=50, age=15, condition="normal"
    assert calculate_dosage(50, 15, "normal") == 100.0
    
    # Test severe condition: weight=50, age=10, condition="severe"
    assert calculate_dosage(50, 10, "severe") == 150.0
    
    # Test severe condition: weight=50, age=12, condition="severe"
    assert calculate_dosage(50, 12, "severe") == 150.0
    
    # Test edge case: invalid weight (0), should raise ValueError
    assert calculate_dosage(0, 10, "normal") == ValueError
    
    # Test edge case: invalid age (-1), should raise ValueError
    assert calculate_dosage(50, -1, "normal") == ValueError
````

### ✏️ Exercise Create a function that processes medical image data:
   - Accept a numpy array representing an image
   - Apply basic image processing (normalization, filtering)
   - Return the processed image
   - Include proper error handling

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter

def process_medical_image(image):
    """
    Process a medical image by normalizing and applying a Gaussian filter.

    Parameters:
        image (numpy.ndarray): Input image as a numpy array.

    Returns:
        numpy.ndarray: Processed image.

    Raises:
        ValueError: If the input is not a numpy array or is empty.
    """
    # Check if the input is a valid numpy array
    if not isinstance(image, np.ndarray):
        raise ValueError("Input must be a numpy array.")
    if image.size == 0:
        raise ValueError("Input image cannot be empty.")

    # Normalize the image to the range [0, 1]
    normalized_image = (image - np.min(image)) / (np.max(image) - np.min(image))

    # Apply a Gaussian filter for smoothing
    filtered_image = gaussian_filter(normalized_image, sigma=1)

    # Return the processed image
    return filtered_image

# Example usage:
# Generate a random example image
image = np.random.rand(256, 256)

# Process the image
processed_image = process_medical_image(image)

# Display the original and processed images
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap="gray")
plt.colorbar()

# Processed image
plt.subplot(1, 2, 2)
plt.title("Processed Image")
plt.imshow(processed_image, cmap="gray")
plt.colorbar()

plt.tight_layout()
plt.show()
````


### ✏️ Exercise: Write a class for managing patient records
   - Include methods for adding, updating, and retrieving records
   - Implement data validation
   - Add methods for basic statistical analysis

````python
class PatientRecords:
    """
    A class to manage patient records with methods for adding, updating,
    retrieving records, and performing basic statistical analysis.
    """

    def __init__(self):
        # Initialize an empty dictionary to store patient records
        self.records = {}

    def add_record(self, patient_id, name, age, condition):
        """
        Add a new patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.
            name (str): Name of the patient.
            age (int): Age of the patient.
            condition (str): Medical condition of the patient.

        Raises:
            ValueError: If the input data is invalid or patient_id already exists.
        """
        if not isinstance(patient_id, str) or not patient_id:
            raise ValueError("Patient ID must be a non-empty string.")
        if patient_id in self.records:
            raise ValueError("Patient ID already exists.")
        if not isinstance(name, str) or not name:
            raise ValueError("Name must be a non-empty string.")
        if not isinstance(age, int) or age <= 0:
            raise ValueError("Age must be a positive integer.")
        if not isinstance(condition, str) or not condition:
            raise ValueError("Condition must be a non-empty string.")

        # Add the record to the dictionary
        self.records[patient_id] = {"name": name, "age": age, "condition": condition}

    def update_record(self, patient_id, name=None, age=None, condition=None):
        """
        Update an existing patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.
            name (str, optional): Updated name of the patient.
            age (int, optional): Updated age of the patient.
            condition (str, optional): Updated medical condition of the patient.

        Raises:
            ValueError: If the patient_id does not exist or input data is invalid.
        """
        if patient_id not in self.records:
            raise ValueError("Patient ID does not exist.")

        # Update fields if provided
        if name is not None:
            if not isinstance(name, str) or not name:
                raise ValueError("Name must be a non-empty string.")
            self.records[patient_id]["name"] = name
        if age is not None:
            if not isinstance(age, int) or age <= 0:
                raise ValueError("Age must be a positive integer.")
            self.records[patient_id]["age"] = age
        if condition is not None:
            if not isinstance(condition, str) or not condition:
                raise ValueError("Condition must be a non-empty string.")
            self.records[patient_id]["condition"] = condition

    def get_record(self, patient_id):
        """
        Retrieve a patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.

        Returns:
            dict: The patient record.

        Raises:
            ValueError: If the patient_id does not exist.
        """
        if patient_id not in self.records:
            raise ValueError("Patient ID does not exist.")
        return self.records[patient_id]

    def average_age(self):
        """
        Calculate the average age of all patients.

        Returns:
            float: The average age of patients, or 0 if no records exist.
        """
        if not self.records:
            return 0
        total_age = sum(record["age"] for record in self.records.values())
        return total_age / len(self.records)

    def count_by_condition(self, condition):
        """
        Count the number of patients with a specific condition.

        Parameters:
            condition (str): The medical condition to count.

        Returns:
            int: The number of patients with the specified condition.
        """
        return sum(1 for record in self.records.values() if record["condition"] == condition)

# Example usage:
# records = PatientRecords()
# records.add_record("001", "Alice", 30, "Diabetes")
# records.add_record("002", "Bob", 40, "Hypertension")
# print(records.get_record("001"))
# records.update_record("001", age=31)
# print(records.average_age())
# print(records.count_by_condition("Diabetes"))
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, widgets

# Sample data
data = {
    "Patient ID": [1, 2, 3, 4, 5],
    "Age": [25, 34, 45, 52, 29],
    "Condition": ["Diabetes", "Hypertension", "Diabetes", "Asthma", "Hypertension"],
    "BMI": [22.5, 27.8, 30.1, 24.3, 26.7],
    "Blood Pressure": [120, 140, 135, 125, 130],
}
df = pd.DataFrame(data)

# Function to display a bar chart of conditions
def plot_condition_distribution():
    condition_counts = df["Condition"].value_counts()
    condition_counts.plot(kind="bar", color="skyblue", edgecolor="black")
    plt.title("Condition Distribution")
    plt.xlabel("Condition")
    plt.ylabel("Count")
    plt.show()

# Function to display a scatter plot of Age vs. BMI
def plot_age_vs_bmi():
    plt.scatter(df["Age"], df["BMI"], color="green", edgecolor="black")
    plt.title("Age vs. BMI")
    plt.xlabel("Age")
    plt.ylabel("BMI")
    plt.grid(True)
    plt.show()

# Function to display a histogram of Blood Pressure
def plot_blood_pressure_histogram():
    plt.hist(df["Blood Pressure"], bins=5, color="orange", edgecolor="black")
    plt.title("Blood Pressure Distribution")
    plt.xlabel("Blood Pressure")
    plt.ylabel("Frequency")
    plt.show()

# Interactive dashboard
def dashboard(plot_type):
    if plot_type == "Condition Distribution":
        plot_condition_distribution()
    elif plot_type == "Age vs. BMI":
        plot_age_vs_bmi()
    elif plot_type == "Blood Pressure Histogram":
        plot_blood_pressure_histogram()

# Dropdown widget for selecting the plot type
interact(dashboard, plot_type=widgets.Dropdown(
    options=["Condition Distribution", "Age vs. BMI", "Blood Pressure Histogram"],
    description="Select Plot:",
))

# Display the default graph (e.g., Condition Distribution) at the end
plot_condition_distribution()
````
